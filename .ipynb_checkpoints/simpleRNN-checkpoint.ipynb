```python
#Simple Keras based experiments for RNNs
# Order = 2, Fibonacci
#
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

np.random.seed(42)

#
# Create simple time series
#
# 1, 1, 2, 3, 5, 8, 13
X = np.zeros([5,1,2])
X[0,0,0] = 1
X[0,0,1] = 1
X[1,0,0] = 1
X[1,0,1] = 2
X[2,0,0] = 2
X[2,0,1] = 3
X[3,0,0] = 3
X[3,0,1] = 5
X[4,0,0] = 5
X[4,0,1] = 8

y = np.zeros([5,1])
y [0,0] = 2
y [1,0] = 3
y [2,0] = 5
y [3,0] = 8
y [4,0] = 13

##print('X = ', X)
##print('y = ', y)

##print('X.shape = ', X.shape)
##print('Y.shape = ', y.shape)

#
# Create simple architecture
#

model = Sequential()
model.add(Flatten(input_shape=X.shape[1:]))
model.add(Dense(1))

# summarize the model
model.summary()

# compile the model
model.compile(loss='mean_squared_error', optimizer='rmsprop', 
              metrics=['accuracy'])
#
# Train
#
from keras.callbacks import ModelCheckpoint   

# train the model
###checkpointer = ModelCheckpoint(filepath='mnist.model.best.hdf5', 
###                               verbose=1, save_best_only=True)
hist = model.fit(X, y, epochs=1000, verbose=0)

print('model.get_weights = ', model.get_weights())
#
# Do some predictions
#

X_test = np.zeros([1,1,2])
X_test[0,0,0] = 5
X_test[0,0,1] = 8

print('Predictions = ', model.predict(X_test))
#
#
#
```

```python
# Logistic recursion
import matplotlib.pyplot as py
#
# define recursion
#
def logrec(val, w):
    return w*val-w*val*val
#
# set up call
#

alpha = 0.0001
w = 4
#
# Generate sequence
#
storeSeq = []
storeSeq.append(alpha)
val = alpha
for i in range(50):
    val = logrec(val,w)
    storeSeq.append(val)
    
#
# Plot
#
plt.plot(storeSeq)
plt.show()
#
#
#
```

```python
#Simple Keras based experiments for RNNs
# even number sequence
#
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, SimpleRNN

np.random.seed(42)

#
# Create simple time series
#
# 
# h2 = 2 + s1
# h3 = h2 + s2 + 1
# ht+1 = ht + 2
# 
# The first dimension is the sample index
# the window size is the second dimension, the third dimension is the dimension of the symbol
#
# For example, we can have 4 outputs, each of which consist of samples from 3 timepoints.
#
X = np.zeros([4,3,1])  
X[0,0,0] = 0
X[0,1,0] = 2
X[0,2,0] = 4
X[1,0,0] = 6
X[1,1,0] = 8
X[1,2,0] = 10
X[2,0,0] = 12
X[2,1,0] = 14
X[2,2,0] = 16
X[3,0,0] = 0
X[3,1,0] = 2
X[3,2,0] = 4

y = np.zeros([4])
y [0] = 6
y [1] = 12
y [2] = 18
y [3] = 6

##print('X = ', X)
##print('y = ', y)

##print('X.shape = ', X.shape)
##print('Y.shape = ', y.shape)

#
# Create simple architecture
#

model = Sequential()
###model.add(Flatten(input_shape=X.shape[1:]))
model.add(SimpleRNN(1, input_shape=(3,1), activation='linear'))

# summarize the model
model.summary()

# compile the model
model.compile(loss='mean_squared_error', optimizer='rmsprop', 
              metrics=['accuracy'])
#
# Train
#
from keras.callbacks import ModelCheckpoint   

# train the model
###checkpointer = ModelCheckpoint(filepath='mnist.model.best.hdf5', 
###                               verbose=1, save_best_only=True)
hist = model.fit(X, y, epochs=3000, verbose=0)

print('model.get_weights = ', model.get_weights())
#
# Do some predictions
#

X_test = np.zeros([1,3,1])
X_test[0,0,0] = 20
X_test[0,1,0] = 22
X_test[0,2,0] = 24

print('Predictions = ', model.predict(X_test))
#
#
#
```

```python
#Simple Keras based experiments for RNNs
# Fibonacci.
#
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, SimpleRNN

np.random.seed(42)

#
# Create simple time series
#
# 1, 1, 2, 3, 5, 8, 13
# 
# h2 = 2 + s1
# h3 = h2 + s2 + 1
# ht+1 = ht + 2
# 
# The first dimension is the sample index
# the window size is the second dimension, the third dimension is the dimension of the symbol
#
# For example, we can have 4 outputs, each of which consist of samples from 2 timepoints.
#
X = np.zeros([2,4,1])  
X[0,0,0] = 1
X[0,1,0] = 1
X[0,2,0] = 2
X[0,3,0] = 3

X[1,0,0] = 1
X[1,1,0] = 2
X[1,2,0] = 3
X[1,3,0] = 5

y = np.zeros([2])
y [0] = 5
y [1] = 8

##print('X = ', X)
##print('y = ', y)

##print('X.shape = ', X.shape)
##print('Y.shape = ', y.shape)

#
# Create simple architecture
#

model = Sequential()
###model.add(Flatten(input_shape=X.shape[1:]))
model.add(SimpleRNN(1, input_shape=(4,1), activation='linear'))

# summarize the model
model.summary()

# compile the model
model.compile(loss='mean_squared_error', optimizer='rmsprop', 
              metrics=['accuracy'])
#
# Train
#
from keras.callbacks import ModelCheckpoint   

# train the model
###checkpointer = ModelCheckpoint(filepath='mnist.model.best.hdf5', 
###                               verbose=1, save_best_only=True)
hist = model.fit(X, y, epochs=3000, verbose=0)

print('model.get_weights = ', model.get_weights())
#
# Do some predictions
#

X_test = np.zeros([1,4,1])
X_test[0,0,0] = 2
X_test[0,1,0] = 3
X_test[0,2,0] = 5
X_test[0,3,0] = 8

print('Predictions = ', model.predict(X_test))
#
#
#
```
